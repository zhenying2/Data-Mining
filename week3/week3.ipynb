## Data Visualization
- Seaborn을 이용한 시각화

```python
import pandas as pd
import seaborn as sns

#data loading
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/데이터마이닝/datas/iris.csv")
sns.scatterplot(x="petal length in cm", y="petal width in cm", data=df, hue="species")
```

```python
sns.lmplot(x="petal length in cm",y="petal width in cm", hue="species", data=df, fit_reg=False, palette="bright",markers=["o","x","v"])
```

```python
penguins=sns.load_dataset("penguins")
print(penguins.head())
sns.displot(data=penguins,x="flipper_length_mm")
```

```python
sns.displot(data=penguins,x="flipper_length_mm",kind="kde")
```

```python
sns.histplot(df["petal length in cm"])
```

```python
sns.histplot(df["petal length in cm"],bins=15)
```

```python
sns.jointplot(x="petal length in cm",y="petal width in cm",data=df,kind="scatter",marginal_kws=dict(bins=10))
```

```python
sns.jointplot(x="petal length in cm",y="petal width in cm",data=df,kind="kde")
```

```python
sns.violinplot(x="species",y="sepal width in cm", data=df)
```

```python
sns.violinplot(x="species",y="petal width in cm",data=df)
```

```python
sns.pairplot(data=df)
```

```python
import pandas as pd
import sqlite3
sqlite_file="/content/drive/MyDrive/Colab Notebooks/데이터마이닝/datas/boston.db"

#Connecting to the database file
conn=sqlite3.connect(sqlite_file)
df=pd.read_sql_query("select * from boston;",conn)
vars_to_plot=["CRIM","AGE","DIS","LSTAT","MEDV"]
sns.pairplot(data=df,vars=vars_to_plot)
```

```python
import pandas as pd
import seaborn as sns

#data loading
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/데이터마이닝/datas/iris.csv")
```

```python
# 색깔 지정
colors = ["#191970","#B8860B", "#228B22"]
sns.set_palette(sns.color_palette(colors))

sns.lmplot(x="sepal width in cm",y="petal width in cm", hue="species", data=df, fit_reg=False,markers=["p","*","d"])
```
